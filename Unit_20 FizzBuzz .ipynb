# Unit20: FIzzBuzz

```python
for i in range(1,101):
    if i % 3 == 0 and i % 5 == 0:  # 제일 앞에 와야한다. 뒤에 있으면 error
        print('FizzBuzz')           # 혹은 if i % 15 == 0 
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
        
for i in range(1,101):
    print('Fizz'*(i%3==0) + 'Buzz' * (i % 5 == 0) or i)
```
